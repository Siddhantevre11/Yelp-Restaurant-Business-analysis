```
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
```

```
from nltk.stem import WordNetLemmatizer 
  
lemmatizer = WordNetLemmatizer() 
```

```
array = []
def text_clean(df):
  for _, row in df.iterrows():
    text = row['text']
    text = text.lower()  #lowercase
    words = nltk.word_tokenize(text)    #tokenize
    new_words= [word for word in words if word.isalpha()]   #Remove punctuations and numbers
    nostop = [words for words in new_words if words not in ENGLISH_STOP_WORDS]    #Remove stopwords
    #nopos = [words for words in nostop if words not in positive_words] 
    lema = [lemmatizer.lemmatize(words) for words in nostop]   #lemmatize words
    array.append(lema)
  return array
```

```
arr_positive = text_clean(tr_positive)
```

```
positive_dict = {}
for sentence_array in arr_positive:
  for raw_word in sentence_array:
    if raw_word not in positive_dict:
        positive_dict[raw_word] = 0 
    positive_dict[raw_word] += 1
```

```
positive_dict
```

```
arr_negative = text_clean(tr_negative)
```

```
negative_dict = {}
for sentence_array in arr_negative:
  for raw_word in sentence_array:
    if raw_word not in negative_dict:
        negative_dict[raw_word] = 0 
    negative_dict[raw_word] += 1
```

```
negative_dict
```

```
def mergeDict(dict1, dict2):
   dict3 = {**dict1, **dict2}
   for key, value in dict3.items():
       if key in dict1 and key in dict2:
               dict3[key] = value + dict1[key]
   return dict3
 
# Merge dictionaries and add values of common keys in a list
dict3 = mergeDict(positive_dict, negative_dict)

print('Dictionary 3 :')
print(dict3)
```
